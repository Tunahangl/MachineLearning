```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Ã–rnek veri seti oluÅŸturma
data = {
    'Ã–ÄŸrenci_ID': [1, 2, 3, 4, 5],
    'Ad': ['Ali', 'AyÅŸe', np.nan, 'Mehmet', 'Zeynep'],
    'Matematik': [85, 90, np.nan, 75, np.nan],
    'Fizik': [78, np.nan, 82, 88, 92],
    'Kimya': [90, 85, 70, np.nan, np.nan],
    'SÄ±nÄ±f': ['10A', '10B', np.nan, '10A', '10B'],
    'KatÄ±lÄ±m': [1, np.nan, 0, 1, np.nan]
}
df = pd.DataFrame(data)

print("ORÄ°JÄ°NAL VERÄ° SETÄ°:")
print(df.to_markdown(index=False))

# 1. SABÄ°T DEÄžERLE DOLDURMA
print("\n1. SABÄ°T DEÄžERLE DOLDURMA:")
# TÃ¼m NaN'larÄ± 0 ile doldur
df_fill_zero = df.fillna(0)
print("\na) TÃ¼m eksikler 0 olarak:")
print(df_fill_zero.to_markdown(index=False))

# SÃ¼tunlara Ã¶zel deÄŸerler
df_fill_custom = df.fillna({
    'Ad': 'Bilinmiyor',
    'Matematik': df['Matematik'].mean(),
    'SÄ±nÄ±f': 'Belirsiz'
})
print("\nb) SÃ¼tunlara Ã¶zel doldurma:")
print(df_fill_custom.to_markdown(index=False))

# 2. Ä°STATÄ°STÄ°KSEL DEÄžERLERLE DOLDURMA
print("\n2. Ä°STATÄ°STÄ°KSEL DEÄžERLERLE DOLDURMA:")
# SayÄ±sal sÃ¼tunlarÄ±n ortalamasÄ±
df_fill_mean = df.fillna(df.mean(numeric_only=True))
print("\na) Ortalama ile doldurma:")
print(df_fill_mean.to_markdown(index=False))

# Medyan ile doldurma
df_fill_median = df.fillna(df.median(numeric_only=True))
print("\nb) Medyan ile doldurma:")
print(df_fill_median.to_markdown(index=False))

# Mod ile doldurma (kategorik veriler iÃ§in)
df_fill_mode = df.fillna(df.mode().iloc[0])
print("\nc) Mod ile doldurma:")
print(df_fill_mode.to_markdown(index=False))

# 3. KOMÅžU DEÄžERLERLE DOLDURMA
print("\n3. KOMÅžU DEÄžERLERLE DOLDURMA:")
# Ã–nceki deÄŸerle doldur (forward fill)
df_ffill = df.fillna(method='ffill')
print("\na) Ã–nceki deÄŸerle doldurma (ffill):")
print(df_ffill.to_markdown(index=False))

# Sonraki deÄŸerle doldur (backward fill)
df_bfill = df.fillna(method='bfill')
print("\nb) Sonraki deÄŸerle doldurma (bfill):")
print(df_bfill.to_markdown(index=False))

# 4. SKLEARN Ä°LE DOLDURMA (GELÄ°ÅžMÄ°Åž YÃ–NTEMLER)
print("\n4. SKLEARN Ä°LE DOLDURMA:")
# SimpleImputer kullanÄ±mÄ±
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df.select_dtypes(include=['number'])), 
                         columns=df.select_dtypes(include=['number']).columns)
print("\na) SimpleImputer (mean) ile sayÄ±sal sÃ¼tunlar:")
print(df_imputed.to_markdown(index=False))

# Kategorik veriler iÃ§in
imputer_cat = SimpleImputer(strategy='most_frequent')
df_imputed_cat = pd.DataFrame(imputer_cat.fit_transform(df.select_dtypes(exclude=['number'])), 
                            columns=df.select_dtypes(exclude=['number']).columns)
print("\nb) SimpleImputer (most_frequent) ile kategorik sÃ¼tunlar:")
print(df_imputed_cat.to_markdown(index=False))

# 5. KARIÅžIK YÃ–NTEMLERLE DOLDURMA
print("\n5. KARIÅžIK YÃ–NTEMLERLE DOLDURMA:")
# SayÄ±sal sÃ¼tunlara ortalama, kategoriklere mod, diÄŸerlerine sabit deÄŸer
df_mixed = df.copy()
df_mixed[df.select_dtypes(include=['number']).columns] = df_mixed.select_dtypes(include=['number']).fillna(df.mean())
df_mixed[df.select_dtypes(exclude=['number']).columns] = df_mixed.select_dtypes(exclude=['number']).fillna(df.mode().iloc[0])
print("\nKarÄ±ÅŸÄ±k doldurma yÃ¶ntemi:")
print(df_mixed.to_markdown(index=False))

# EKSÄ°K VERÄ° ANALÄ°Z RAPORU
print("\nEKSÄ°K VERÄ° ANALÄ°Z RAPORU:")
print("Eksik veri sayÄ±larÄ±:")
print(df.isnull().sum().to_markdown())
print("\nEksik veri yÃ¼zdeleri:")
print((df.isnull().sum()/len(df)*100).round(2).to_markdown())

"""
Ã‡IKTI AÃ‡IKLAMALARI:

1. Sabit DeÄŸerle:
   - TÃ¼m NaN'lar 0 yapÄ±ldÄ±
   - Ã–zel doldurmada Ad -> 'Bilinmiyor', Matematik -> ortalama (83.33)

2. Ä°statistiksel:
   - Ortalama: Matematik (85+90+75)/3 â‰ˆ 83.33
   - Medyan: Matematik [75,85,90] -> 85
   - Mod: SÄ±nÄ±f '10A' ve '10B' 2'ÅŸer kez (ilk deÄŸer '10A' alÄ±ndÄ±)

3. KomÅŸu DeÄŸerler:
   - ffill: 3. satÄ±r Ad -> 'AyÅŸe' (Ã¼stÃ¼ndeki)
   - bfill: 2. satÄ±r Fizik -> 82 (altÄ±ndaki)

4. Sklearn:
   - SimpleImputer sayÄ±sal sÃ¼tunlara ortalama uyguladÄ±
   - Kategorik sÃ¼tunlara en sÄ±k deÄŸer (mode) atandÄ±

5. KarÄ±ÅŸÄ±k:
   - SayÄ±sal sÃ¼tunlar ortalama ile
   - Kategorik sÃ¼tunlar mod ile dolduruldu
"""
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Ã–rnek veri seti oluÅŸturma
data = {
    'Åžehir': ['Ä°stanbul', 'Ankara', np.nan, 'Ä°zmir', np.nan, 'Bursa', np.nan],
    'Renk': ['KÄ±rmÄ±zÄ±', np.nan, 'Mavi', np.nan, 'YeÅŸil', 'KÄ±rmÄ±zÄ±', np.nan],
    'ÃœrÃ¼n_Kategori': ['Elektronik', 'Giyim', np.nan, np.nan, 'Giyim', 'Elektronik', 'Mobilya'],
    'MÃ¼ÅŸteri_Tipi': [np.nan, 'Bireysel', 'Kurumsal', np.nan, 'Bireysel', np.nan, 'Kurumsal']
}
df = pd.DataFrame(data)

print("ORÄ°JÄ°NAL VERÄ° SETÄ°:")
print(df.to_markdown(index=False))

# 1. SABÄ°T DEÄžERLE DOLDURMA
print("\n1. SABÄ°T DEÄžERLE DOLDURMA:")

# TÃ¼m kategorik sÃ¼tunlarda eksikleri 'Bilinmiyor' ile doldur
df_sabit = df.copy()
kategorik_sutunlar = ['Åžehir', 'Renk', 'ÃœrÃ¼n_Kategori', 'MÃ¼ÅŸteri_Tipi']
df_sabit[kategorik_sutunlar] = df_sabit[kategorik_sutunlar].fillna('Bilinmiyor')

print("\na) TÃ¼m kategoriklerde 'Bilinmiyor' ile doldurma:")
print(df_sabit.to_markdown(index=False))

# SÃ¼tunlara Ã¶zel sabit deÄŸerler
df_sabit_ozel = df.copy()
df_sabit_ozel['Åžehir'] = df_sabit_ozel['Åžehir'].fillna('Belirsiz')
df_sabit_ozel['MÃ¼ÅŸteri_Tipi'] = df_sabit_ozel['MÃ¼ÅŸteri_Tipi'].fillna('DiÄŸer')

print("\nb) SÃ¼tunlara Ã¶zel sabit deÄŸerlerle doldurma:")
print(df_sabit_ozel.to_markdown(index=False))

# 2. MOD DEÄžERÄ°YLE DOLDURMA (EN SIK TEKRARLANAN)
print("\n2. MOD DEÄžERÄ°YLE DOLDURMA:")

df_mod = df.copy()

# Her sÃ¼tunun mod deÄŸerini ayrÄ± ayrÄ± hesaplayarak doldurma
for sutun in kategorik_sutunlar:
    mod_deger = df_mod[sutun].mode()[0]  # En sÄ±k gÃ¶rÃ¼len deÄŸer
    df_mod[sutun] = df_mod[sutun].fillna(mod_deger)
    print(f"{sutun} sÃ¼tunu mod deÄŸeri: {mod_deger}")

print("\nMod deÄŸerleriyle doldurulmuÅŸ veri seti:")
print(df_mod.to_markdown(index=False))

# 3. SKLEARN Ä°LE DOLDURMA
print("\n3. SKLEARN Ä°LE OTOMATÄ°K DOLDURMA:")

imputer = SimpleImputer(strategy='most_frequent')
df_sklearn = pd.DataFrame(imputer.fit_transform(df[kategorik_sutunlar]), 
                         columns=kategorik_sutunlar)

print("SimpleImputer (most_frequent) ile doldurulmuÅŸ veri:")
print(df_sklearn.to_markdown(index=False))

# 4. KOÅžULLU DOLDURMA
print("\n4. KOÅžULLU DOLDURMA Ã–RNEÄžÄ°:")

df_kosullu = df.copy()

# Sadece belirli koÅŸullarda doldurma
df_kosullu['Åžehir'] = np.where(
    (df_kosullu['Åžehir'].isna()) & (df_kosullu['ÃœrÃ¼n_Kategori'] == 'Elektronik'),
    'Ä°stanbul',  # KoÅŸul saÄŸlanÄ±yorsa bu deÄŸer
    df_kosullu['Åžehir'].fillna('DiÄŸer')  # SaÄŸlanmÄ±yorsa 'DiÄŸer' ile doldur
)

print("KoÅŸullu doldurma sonrasÄ±:")
print(df_kosullu.to_markdown(index=False))

# EKSÄ°K VERÄ° ANALÄ°Z RAPORU
print("\nEKSÄ°K VERÄ° ANALÄ°Z RAPORU:")
print("Eksik veri sayÄ±larÄ±:")
print(df.isnull().sum().to_markdown())
print("\nEksik veri yÃ¼zdeleri:")
print((df.isnull().sum()/len(df)*100).round(2).to_markdown())

"""
Ã‡IKTI AÃ‡IKLAMALARI:

1. Sabit DeÄŸerle Doldurma:
   - TÃ¼m kategorik sÃ¼tunlardaki NaN'lar 'Bilinmiyor' yapÄ±ldÄ±
   - Ã–zel versiyonda Åžehir->'Belirsiz', MÃ¼ÅŸteri_Tipi->'DiÄŸer' yapÄ±ldÄ±

2. Mod DeÄŸeriyle Doldurma:
   - Åžehir modu: Ä°stanbul (2 kez)
   - Renk modu: KÄ±rmÄ±zÄ± (2 kez)
   - ÃœrÃ¼n_Kategori modu: Elektronik (2 kez)
   - MÃ¼ÅŸteri_Tipi modu: Bireysel (2 kez)

3. Sklearn:
   - SimpleImputer otomatik olarak en sÄ±k deÄŸerleri hesaplayÄ±p doldurdu

4. KoÅŸullu Doldurma:
   - ÃœrÃ¼n_Kategori 'Elektronik' olan ve Åžehir NaN olanlar 'Ä°stanbul' yapÄ±ldÄ±
   - DiÄŸer Åžehir NaN'larÄ± 'DiÄŸer' yapÄ±ldÄ±
"""
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore


veri = pd.DataFrame({
    "sayi": [10, 12, 14, 15, 13, 12, 14, 13, 16, 200]
})


Q1 = veri["sayi"].quantile(0.25)
Q3 = veri["sayi"].quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

veri["iqr_aykiri"] = ((veri["sayi"] < alt_sinir) | (veri["sayi"] > ust_sinir)).astype(int)


veri["zscore"] = zscore(veri["sayi"])
veri["zscore_aykiri"] = (np.abs(veri["zscore"]) > 2.5).astype(int)


veri["ikisi_de_aykiri"] = ((veri["iqr_aykiri"] == 1) & (veri["zscore_aykiri"] == 1)).astype(int)


print(veri)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
import matplotlib.pyplot as plt


veri = pd.DataFrame({
    "sayi": [10, 12, 14, 15, 13, 12, 14, 13, 16, 200]
})


Q1 = veri["sayi"].quantile(0.25)
Q3 = veri["sayi"].quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

veri["iqr_aykiri"] = ((veri["sayi"] < alt_sinir) | (veri["sayi"] > ust_sinir)).astype(int)


veri["zscore"] = zscore(veri["sayi"])
veri["zscore_aykiri"] = (np.abs(veri["zscore"]) > 2.5).astype(int)


print(veri)


plt.figure(figsize=(10,5))


plt.scatter(veri.index, veri["sayi"], label="Veri", color='blue')


plt.scatter(veri[veri["iqr_aykiri"]==1].index, veri[veri["iqr_aykiri"]==1]["sayi"],
            label="IQR AykÄ±rÄ±", color='red', s=100, marker='X')


plt.scatter(veri[veri["zscore_aykiri"]==1].index, veri[veri["zscore_aykiri"]==1]["sayi"],
            label="Z-score AykÄ±rÄ±", color='green', s=100, marker='P')

plt.title("IQR ve Z-Score AykÄ±rÄ± DeÄŸerler")
plt.xlabel("Index")
plt.ylabel("SayÄ±")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
import matplotlib.pyplot as plt


veri = pd.DataFrame({
    "degisken": [5, 7, 8, 6, 5, 9, 120, 6, 7, 300, 8, 7, 5, 6, 250]
})


veri["log_transform"] = np.log1p(veri["degisken"])  


veri["sqrt_transform"] = np.sqrt(veri["degisken"])


veri["boxcox_transform"], lam = boxcox(veri["degisken"])
print(f"Box-Cox Lambda deÄŸeri: {lam:.4f}")


Q1 = veri["boxcox_transform"].quantile(0.25)
Q3 = veri["boxcox_transform"].quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR


veri_temiz = veri[(veri["boxcox_transform"] >= alt_sinir) & (veri["boxcox_transform"] <= ust_sinir)]


print("Orijinal Veri:\n", veri["degisken"])
print("\nBox-Cox DÃ¶nÃ¼ÅŸÃ¼mÃ¼ SonrasÄ± IQR ile TemizlenmiÅŸ Veri:\n", veri_temiz["degisken"])


plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Orijinal Veri")
plt.boxplot(veri["degisken"])
plt.grid(True)

plt.subplot(1,2,2)
plt.title("Box-Cox + IQR TemizlenmiÅŸ Veri")
plt.boxplot(veri_temiz["degisken"])
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from scipy.stats import zscore


california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df["target"] = california.target  # MedHouseVal


for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):#bu kod satÄ±rÄ± sayesinde sayÄ±sal olmayan sÃ¼tunlarÄ± iqr denetiminden uzak tutarÄ±z.        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        alt = Q1 - 1.5 * IQR
        ust = Q3 + 1.5 * IQR
        df[f"{col}_iqr_outlier"] = ((df[col] < alt) | (df[col] > ust)).astype(int)


z_df = df.select_dtypes(include=[np.number]).copy()
z_scores = z_df.apply(zscore)
z_outliers = (np.abs(z_scores) > 3).astype(int)
z_outliers.columns = [col + "_z_outlier" for col in z_outliers.columns]
df = pd.concat([df, z_outliers], axis=1)




X = df[california.feature_names]
y = df["target"]


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


model = DecisionTreeRegressor(max_depth=6, random_state=0)
model.fit(x_train_scaled, y_train)
y_pred = model.predict(x_test_scaled)


mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("RMSE (kÃ¶k ortalama kare hata):", rmse)

""""# 9. Bonus: AykÄ±rÄ±lar ne kadar etkiliyor?
toplam_aykiri = df[[col for col in df.columns if "outlier" in col]].sum().sum()
print("Toplam aykÄ±rÄ± deÄŸer tespiti (IQR+Z):", toplam_aykiri)

# df zaten iÃ§inde aykÄ±rÄ± deÄŸer sÃ¼tunlarÄ± var, isimlerinde "_outlier" geÃ§iyor

# 1. SatÄ±r bazÄ±nda toplam aykÄ±rÄ± sayÄ±sÄ± (her satÄ±rda kaÃ§ aykÄ±rÄ± deÄŸer var)
df["toplam_aykiri_satir"] = df[[col for col in df.columns if "_outlier" in col]].sum(axis=1)

# 2. AykÄ±rÄ± deÄŸer sayÄ±sÄ±na gÃ¶re satÄ±r sayÄ±sÄ±
aykiri_satir_sayisi = df["toplam_aykiri_satir"].value_counts().sort_index()
print("SatÄ±r bazÄ±nda aykÄ±rÄ± deÄŸer sayÄ±sÄ± daÄŸÄ±lÄ±mÄ±:")
print(aykiri_satir_sayisi)

# 3. SÃ¼tun bazÄ±nda toplam aykÄ±rÄ± sayÄ±sÄ± (her Ã¶zellikte kaÃ§ aykÄ±rÄ± gÃ¶zlem var)
outlier_cols = [col for col in df.columns if "_outlier" in col]
aykiri_sutun_sayilari = df[outlier_cols].sum().sort_values(ascending=False)
print("\nSÃ¼tun bazÄ±nda aykÄ±rÄ± deÄŸer sayÄ±larÄ±:")
print(aykiri_sutun_sayilari)
```
