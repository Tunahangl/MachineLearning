```python
import pandas as pd
import numpy as np

# Örnek veri seti oluşturma
data = {
    'Öğrenci_ID': [1, 2, 3, 4, 5, 6],
    'Ad': ['Ali', 'Ayşe', np.nan, 'Mehmet', 'Zeynep', 'Fatma'],
    'Matematik': [85, 90, np.nan, 75, np.nan, 95],
    'Fizik': [78, np.nan, np.nan, 82, 88, 92],
    'Kimya': [90, 85, 70, np.nan, np.nan, np.nan],
    'Sınıf': ['10A', '10B', '10A', '10C', np.nan, '10B']
}
df = pd.DataFrame(data)

print("ORİJİNAL VERİ SETİ:")
print(df.to_markdown(index=False))

# 1. Satır Bazında Silme (Eksik olan tüm satırları siler)
print("\n1. SATIR BAZINDA SİLME (axis=0):")
satir_sil = df.dropna(axis=0)
print(satir_sil.to_markdown(index=False))

# 2. Sütun Bazında Silme (Eksik olan tüm sütunları siler)
print("\n2. SÜTUN BAZINDA SİLME (axis=1):")
sutun_sil = df.dropna(axis=1)
print(sutun_sil.to_markdown(index=False))

# 3. Threshold'a Göre Silme (Belirli bir doluluk oranına göre)
print("\n3. THRESHOLD'A GÖRE SİLME (thresh=4):")
# Her satırda en az 4 dolu değer varsa tutar
thresh_sil = df.dropna(thresh=4)
print(thresh_sil.to_markdown(index=False))

# 4. Belirli Sütunlarda Silme
print("\n4. BELİRLİ SÜTUNLARDA SİLME (subset=['Ad', 'Matematik']):")
# Sadece 'Ad' ve 'Matematik' sütunlarında eksik olan satırları siler
subset_sil = df.dropna(subset=['Ad', 'Matematik'])
print(subset_sil.to_markdown(index=False))

# 5. Dinamik Threshold Hesaplama
print("\n5. DİNAMİK THRESHOLD (%70 doluluk):")
# Sütunlarda %70'ten fazla dolu veri olanları tutar
thresh_dynamic = df.dropna(thresh=len(df)*0.7, axis=1)
print(thresh_dynamic.to_markdown(index=False))

# 6. Eksik Veri Analizi Raporu
print("\n6. EKSİK VERİ ANALİZ RAPORU:")
print("Eksik veri sayıları:")
print(df.isnull().sum().to_markdown())
print("\nEksik veri yüzdeleri:")
print((df.isnull().sum()/len(df)*100).to_markdown())

"""
ÇIKTI AÇIKLAMALARI:

1. Satır Bazında Silme:
   - Sadece 6. satır (Fatma) tüm sütunlarda dolu olduğu için kaldı

2. Sütun Bazında Silme:
   - Sadece Öğrenci_ID sütunu tamamen dolu olduğu için kaldı

3. Threshold'a Göre Silme:
   - 1, 2, 4 ve 6. satırlar en az 4 dolu değere sahip

4. Belirli Sütunlarda Silme:
   - Sadece Ad ve Matematik sütunlarında eksik olmayan satırlar kaldı

5. Dinamik Threshold:
   - Öğrenci_ID, Ad ve Sınıf sütunları %70'ten fazla dolu olduğu için kaldı
"""
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Örnek veri seti oluşturma
data = {
    'Öğrenci_ID': [1, 2, 3, 4, 5],
    'Ad': ['Ali', 'Ayşe', np.nan, 'Mehmet', 'Zeynep'],
    'Matematik': [85, 90, np.nan, 75, np.nan],
    'Fizik': [78, np.nan, 82, 88, 92],
    'Kimya': [90, 85, 70, np.nan, np.nan],
    'Sınıf': ['10A', '10B', np.nan, '10A', '10B'],
    'Katılım': [1, np.nan, 0, 1, np.nan]
}
df = pd.DataFrame(data)

print("ORİJİNAL VERİ SETİ:")
print(df.to_markdown(index=False))

# 1. SABİT DEĞERLE DOLDURMA
print("\n1. SABİT DEĞERLE DOLDURMA:")
# Tüm NaN'ları 0 ile doldur
df_fill_zero = df.fillna(0)
print("\na) Tüm eksikler 0 olarak:")
print(df_fill_zero.to_markdown(index=False))

# Sütunlara özel değerler
df_fill_custom = df.fillna({
    'Ad': 'Bilinmiyor',
    'Matematik': df['Matematik'].mean(),
    'Sınıf': 'Belirsiz'
})
print("\nb) Sütunlara özel doldurma:")
print(df_fill_custom.to_markdown(index=False))

# 2. İSTATİSTİKSEL DEĞERLERLE DOLDURMA
print("\n2. İSTATİSTİKSEL DEĞERLERLE DOLDURMA:")
# Sayısal sütunların ortalaması
df_fill_mean = df.fillna(df.mean(numeric_only=True))
print("\na) Ortalama ile doldurma:")
print(df_fill_mean.to_markdown(index=False))

# Medyan ile doldurma
df_fill_median = df.fillna(df.median(numeric_only=True))
print("\nb) Medyan ile doldurma:")
print(df_fill_median.to_markdown(index=False))

# Mod ile doldurma (kategorik veriler için)
df_fill_mode = df.fillna(df.mode().iloc[0])
print("\nc) Mod ile doldurma:")
print(df_fill_mode.to_markdown(index=False))

# 3. KOMŞU DEĞERLERLE DOLDURMA
print("\n3. KOMŞU DEĞERLERLE DOLDURMA:")
# Önceki değerle doldur (forward fill)
df_ffill = df.fillna(method='ffill')
print("\na) Önceki değerle doldurma (ffill):")
print(df_ffill.to_markdown(index=False))

# Sonraki değerle doldur (backward fill)
df_bfill = df.fillna(method='bfill')
print("\nb) Sonraki değerle doldurma (bfill):")
print(df_bfill.to_markdown(index=False))

# 4. SKLEARN İLE DOLDURMA (GELİŞMİŞ YÖNTEMLER)
print("\n4. SKLEARN İLE DOLDURMA:")
# SimpleImputer kullanımı
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df.select_dtypes(include=['number'])), 
                         columns=df.select_dtypes(include=['number']).columns)
print("\na) SimpleImputer (mean) ile sayısal sütunlar:")
print(df_imputed.to_markdown(index=False))

# Kategorik veriler için
imputer_cat = SimpleImputer(strategy='most_frequent')
df_imputed_cat = pd.DataFrame(imputer_cat.fit_transform(df.select_dtypes(exclude=['number'])), 
                            columns=df.select_dtypes(exclude=['number']).columns)
print("\nb) SimpleImputer (most_frequent) ile kategorik sütunlar:")
print(df_imputed_cat.to_markdown(index=False))

# 5. KARIŞIK YÖNTEMLERLE DOLDURMA
print("\n5. KARIŞIK YÖNTEMLERLE DOLDURMA:")
# Sayısal sütunlara ortalama, kategoriklere mod, diğerlerine sabit değer
df_mixed = df.copy()
df_mixed[df.select_dtypes(include=['number']).columns] = df_mixed.select_dtypes(include=['number']).fillna(df.mean())
df_mixed[df.select_dtypes(exclude=['number']).columns] = df_mixed.select_dtypes(exclude=['number']).fillna(df.mode().iloc[0])
print("\nKarışık doldurma yöntemi:")
print(df_mixed.to_markdown(index=False))

# EKSİK VERİ ANALİZ RAPORU
print("\nEKSİK VERİ ANALİZ RAPORU:")
print("Eksik veri sayıları:")
print(df.isnull().sum().to_markdown())
print("\nEksik veri yüzdeleri:")
print((df.isnull().sum()/len(df)*100).round(2).to_markdown())

"""
ÇIKTI AÇIKLAMALARI:

1. Sabit Değerle:
   - Tüm NaN'lar 0 yapıldı
   - Özel doldurmada Ad -> 'Bilinmiyor', Matematik -> ortalama (83.33)

2. İstatistiksel:
   - Ortalama: Matematik (85+90+75)/3 ≈ 83.33
   - Medyan: Matematik [75,85,90] -> 85
   - Mod: Sınıf '10A' ve '10B' 2'şer kez (ilk değer '10A' alındı)

3. Komşu Değerler:
   - ffill: 3. satır Ad -> 'Ayşe' (üstündeki)
   - bfill: 2. satır Fizik -> 82 (altındaki)

4. Sklearn:
   - SimpleImputer sayısal sütunlara ortalama uyguladı
   - Kategorik sütunlara en sık değer (mode) atandı

5. Karışık:
   - Sayısal sütunlar ortalama ile
   - Kategorik sütunlar mod ile dolduruldu
"""
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Örnek veri seti oluşturma
data = {
    'Şehir': ['İstanbul', 'Ankara', np.nan, 'İzmir', np.nan, 'Bursa', np.nan],
    'Renk': ['Kırmızı', np.nan, 'Mavi', np.nan, 'Yeşil', 'Kırmızı', np.nan],
    'Ürün_Kategori': ['Elektronik', 'Giyim', np.nan, np.nan, 'Giyim', 'Elektronik', 'Mobilya'],
    'Müşteri_Tipi': [np.nan, 'Bireysel', 'Kurumsal', np.nan, 'Bireysel', np.nan, 'Kurumsal']
}
df = pd.DataFrame(data)

print("ORİJİNAL VERİ SETİ:")
print(df.to_markdown(index=False))

# 1. SABİT DEĞERLE DOLDURMA
print("\n1. SABİT DEĞERLE DOLDURMA:")

# Tüm kategorik sütunlarda eksikleri 'Bilinmiyor' ile doldur
df_sabit = df.copy()
kategorik_sutunlar = ['Şehir', 'Renk', 'Ürün_Kategori', 'Müşteri_Tipi']
df_sabit[kategorik_sutunlar] = df_sabit[kategorik_sutunlar].fillna('Bilinmiyor')

print("\na) Tüm kategoriklerde 'Bilinmiyor' ile doldurma:")
print(df_sabit.to_markdown(index=False))

# Sütunlara özel sabit değerler
df_sabit_ozel = df.copy()
df_sabit_ozel['Şehir'] = df_sabit_ozel['Şehir'].fillna('Belirsiz')
df_sabit_ozel['Müşteri_Tipi'] = df_sabit_ozel['Müşteri_Tipi'].fillna('Diğer')

print("\nb) Sütunlara özel sabit değerlerle doldurma:")
print(df_sabit_ozel.to_markdown(index=False))

# 2. MOD DEĞERİYLE DOLDURMA (EN SIK TEKRARLANAN)
print("\n2. MOD DEĞERİYLE DOLDURMA:")

df_mod = df.copy()

# Her sütunun mod değerini ayrı ayrı hesaplayarak doldurma
for sutun in kategorik_sutunlar:
    mod_deger = df_mod[sutun].mode()[0]  # En sık görülen değer
    df_mod[sutun] = df_mod[sutun].fillna(mod_deger)
    print(f"{sutun} sütunu mod değeri: {mod_deger}")

print("\nMod değerleriyle doldurulmuş veri seti:")
print(df_mod.to_markdown(index=False))

# 3. SKLEARN İLE DOLDURMA
print("\n3. SKLEARN İLE OTOMATİK DOLDURMA:")

imputer = SimpleImputer(strategy='most_frequent')
df_sklearn = pd.DataFrame(imputer.fit_transform(df[kategorik_sutunlar]), 
                         columns=kategorik_sutunlar)

print("SimpleImputer (most_frequent) ile doldurulmuş veri:")
print(df_sklearn.to_markdown(index=False))

# 4. KOŞULLU DOLDURMA
print("\n4. KOŞULLU DOLDURMA ÖRNEĞİ:")

df_kosullu = df.copy()

# Sadece belirli koşullarda doldurma
df_kosullu['Şehir'] = np.where(
    (df_kosullu['Şehir'].isna()) & (df_kosullu['Ürün_Kategori'] == 'Elektronik'),
    'İstanbul',  # Koşul sağlanıyorsa bu değer
    df_kosullu['Şehir'].fillna('Diğer')  # Sağlanmıyorsa 'Diğer' ile doldur
)

print("Koşullu doldurma sonrası:")
print(df_kosullu.to_markdown(index=False))

# EKSİK VERİ ANALİZ RAPORU
print("\nEKSİK VERİ ANALİZ RAPORU:")
print("Eksik veri sayıları:")
print(df.isnull().sum().to_markdown())
print("\nEksik veri yüzdeleri:")
print((df.isnull().sum()/len(df)*100).round(2).to_markdown())

"""
ÇIKTI AÇIKLAMALARI:

1. Sabit Değerle Doldurma:
   - Tüm kategorik sütunlardaki NaN'lar 'Bilinmiyor' yapıldı
   - Özel versiyonda Şehir->'Belirsiz', Müşteri_Tipi->'Diğer' yapıldı

2. Mod Değeriyle Doldurma:
   - Şehir modu: İstanbul (2 kez)
   - Renk modu: Kırmızı (2 kez)
   - Ürün_Kategori modu: Elektronik (2 kez)
   - Müşteri_Tipi modu: Bireysel (2 kez)

3. Sklearn:
   - SimpleImputer otomatik olarak en sık değerleri hesaplayıp doldurdu

4. Koşullu Doldurma:
   - Ürün_Kategori 'Elektronik' olan ve Şehir NaN olanlar 'İstanbul' yapıldı
   - Diğer Şehir NaN'ları 'Diğer' yapıldı
"""
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore


veri = pd.DataFrame({
    "sayi": [10, 12, 14, 15, 13, 12, 14, 13, 16, 200]
})


Q1 = veri["sayi"].quantile(0.25)
Q3 = veri["sayi"].quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

veri["iqr_aykiri"] = ((veri["sayi"] < alt_sinir) | (veri["sayi"] > ust_sinir)).astype(int)


veri["zscore"] = zscore(veri["sayi"])
veri["zscore_aykiri"] = (np.abs(veri["zscore"]) > 2.5).astype(int)


veri["ikisi_de_aykiri"] = ((veri["iqr_aykiri"] == 1) & (veri["zscore_aykiri"] == 1)).astype(int)


print(veri)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
import matplotlib.pyplot as plt


veri = pd.DataFrame({
    "sayi": [10, 12, 14, 15, 13, 12, 14, 13, 16, 200]
})


Q1 = veri["sayi"].quantile(0.25)
Q3 = veri["sayi"].quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

veri["iqr_aykiri"] = ((veri["sayi"] < alt_sinir) | (veri["sayi"] > ust_sinir)).astype(int)


veri["zscore"] = zscore(veri["sayi"])
veri["zscore_aykiri"] = (np.abs(veri["zscore"]) > 2.5).astype(int)


print(veri)


plt.figure(figsize=(10,5))


plt.scatter(veri.index, veri["sayi"], label="Veri", color='blue')


plt.scatter(veri[veri["iqr_aykiri"]==1].index, veri[veri["iqr_aykiri"]==1]["sayi"],
            label="IQR Aykırı", color='red', s=100, marker='X')


plt.scatter(veri[veri["zscore_aykiri"]==1].index, veri[veri["zscore_aykiri"]==1]["sayi"],
            label="Z-score Aykırı", color='green', s=100, marker='P')

plt.title("IQR ve Z-Score Aykırı Değerler")
plt.xlabel("Index")
plt.ylabel("Sayı")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
import matplotlib.pyplot as plt


veri = pd.DataFrame({
    "degisken": [5, 7, 8, 6, 5, 9, 120, 6, 7, 300, 8, 7, 5, 6, 250]
})


veri["log_transform"] = np.log1p(veri["degisken"])  


veri["sqrt_transform"] = np.sqrt(veri["degisken"])


veri["boxcox_transform"], lam = boxcox(veri["degisken"])
print(f"Box-Cox Lambda değeri: {lam:.4f}")


Q1 = veri["boxcox_transform"].quantile(0.25)
Q3 = veri["boxcox_transform"].quantile(0.75)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR


veri_temiz = veri[(veri["boxcox_transform"] >= alt_sinir) & (veri["boxcox_transform"] <= ust_sinir)]


print("Orijinal Veri:\n", veri["degisken"])
print("\nBox-Cox Dönüşümü Sonrası IQR ile Temizlenmiş Veri:\n", veri_temiz["degisken"])


plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Orijinal Veri")
plt.boxplot(veri["degisken"])
plt.grid(True)

plt.subplot(1,2,2)
plt.title("Box-Cox + IQR Temizlenmiş Veri")
plt.boxplot(veri_temiz["degisken"])
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from scipy.stats import zscore


california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df["target"] = california.target  # MedHouseVal


for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):#bu kod satırı sayesinde sayısal olmayan sütunları iqr denetiminden uzak tutarız.        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        alt = Q1 - 1.5 * IQR
        ust = Q3 + 1.5 * IQR
        df[f"{col}_iqr_outlier"] = ((df[col] < alt) | (df[col] > ust)).astype(int)


z_df = df.select_dtypes(include=[np.number]).copy()
z_scores = z_df.apply(zscore)
z_outliers = (np.abs(z_scores) > 3).astype(int)
z_outliers.columns = [col + "_z_outlier" for col in z_outliers.columns]
df = pd.concat([df, z_outliers], axis=1)




X = df[california.feature_names]
y = df["target"]


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


model = DecisionTreeRegressor(max_depth=6, random_state=0)
model.fit(x_train_scaled, y_train)
y_pred = model.predict(x_test_scaled)


mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("RMSE (kök ortalama kare hata):", rmse)

""""# 9. Bonus: Aykırılar ne kadar etkiliyor?
toplam_aykiri = df[[col for col in df.columns if "outlier" in col]].sum().sum()
print("Toplam aykırı değer tespiti (IQR+Z):", toplam_aykiri)

# df zaten içinde aykırı değer sütunları var, isimlerinde "_outlier" geçiyor

# 1. Satır bazında toplam aykırı sayısı (her satırda kaç aykırı değer var)
df["toplam_aykiri_satir"] = df[[col for col in df.columns if "_outlier" in col]].sum(axis=1)

# 2. Aykırı değer sayısına göre satır sayısı
aykiri_satir_sayisi = df["toplam_aykiri_satir"].value_counts().sort_index()
print("Satır bazında aykırı değer sayısı dağılımı:")
print(aykiri_satir_sayisi)

# 3. Sütun bazında toplam aykırı sayısı (her özellikte kaç aykırı gözlem var)
outlier_cols = [col for col in df.columns if "_outlier" in col]
aykiri_sutun_sayilari = df[outlier_cols].sum().sort_values(ascending=False)
print("\nSütun bazında aykırı değer sayıları:")
print(aykiri_sutun_sayilari)
```
